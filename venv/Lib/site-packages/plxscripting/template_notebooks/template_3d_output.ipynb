# Intro

When a new notebook is created through PLAXIS 3D Output, you have two objects available. One of them is the "s_o" object which represents the application server which can be accessed as follows:

```python
s_o  # The application server
```

You can also access the "g_o" object which represents the global environment:

```python
g_o  # the global environment
```

You can also observe the current project opened in PLAXIS 3D Output:

```python
current_project = str(g_o.GeneralInfo.Filename)
print(current_project)
```

Now here's an example session with both of the previously mentioned objects, but don't forget to activate the remote scripting server in PLAXIS 3D Output:

```python
s_o.open(current_project)
result_type = g_o.ResultTypes.Soil.Utot
results = g_o.getresults(g_o.Phases[-1], result_type, "node")
max_result = g_o.filter(results, "max")
g_o.echo(max_result)
```
